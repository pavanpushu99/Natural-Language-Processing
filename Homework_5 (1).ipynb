```python
!pip install fasttext
import fasttext
import time
start_time = time.time()
filen = 'medical_dataset.csv'
path = '/content/drive/MyDrive/'
model = fasttext.train_unsupervised(path + filen, model='skipgram')
end_time = time.time()
print(f"training time took {round(end_time - start_time, 3)} secs")
```

```python
print(model.words)
print(len(model['covid-19']) )
```

```python
from sklearn.metrics.pairwise import cosine_similarity
print(cosine_similarity([model['covid-19']], [model['sugar']]))
print(cosine_similarity([model['python']], [model['java']]))
model.get_nearest_neighbors("jupyter", k=6)
```
